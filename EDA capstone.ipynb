*Handling the outliers

```python
sns.boxplot(df_new['price'])
```

```python
ax= sns.violinplot(x='neighbourhood_group',y='price',data=df_new)
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
def outlier_treatment(datacolumn):
 sorted(datacolumn)
 Q1,Q3 = np.percentile(datacolumn , [25,75])
 IQR = Q3 - Q1
 lower_range = Q1 - (1.5 * IQR)
 upper_range = Q3 + (1.5 * IQR)
 
 return lower_range,upper_range
```

```python
lower_bound,upper_bound = outlier_treatment(df1['price'])
```

```python
#df1[(df1.price < lower_bound) | (df1.price > upper_bound)] 
df_new2 = df1[(df1.price>lower_bound) & (df1.price<upper_bound)]
```

```python
sns.distplot(df_new2['price'])
```

```python
ax= sns.violinplot(x='neighbourhood_group',y='price',data=df_new2)
```

We clearly shoen good distribution after outlier deletion
